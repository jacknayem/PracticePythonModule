```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
labels = np.array(['A', 'B','C', 'D'])
plt.pie(y, labels = labels)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
labels = np.array(['A', 'B','C', 'D'])
plt.pie(y, labels = labels, explode=[0.08, 0, 0, 0])
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
labels = np.array(['A', 'B','C', 'D'])
plt.pie(y, labels = labels, explode=[0.08, 0, 0, 0], shadow=True)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
labels = np.array(['A', 'B','C', 'D'])
plt.pie(y, labels = labels, explode=[0.08, 0, 0, 0], shadow=True)
plt.legend(title='List')
plt.show()
```
