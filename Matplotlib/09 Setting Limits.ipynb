```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,10)
fig = plt.figure()
ex = fig.add_axes([0,0,1,1])
ex.plot(x, np.exp(x))
plt.show()
```

```python
x = np.arange(1,10)
fig = plt.figure()
ex = fig.add_axes([0,0,1,1])
ex.plot(x, np.exp(x))
ex.set_xlim(1,8.5)
ex.set_ylim(0, 6000)
plt.show()
```
