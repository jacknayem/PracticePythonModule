```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
ax.set_title("Sine")
ax.set_xlabel('Angle')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
ax.set_title("Sine")
ax.set_xlabel('Angle')
ax.set_xticks([0,2,4,6])
ax.set_xticklabels(['zero', 'two', 'four', 'six'])
ax.set_yticks([-1, 0, 1])
plt.show()
```
