```python
import matplotlib.pyplot as plt
import numpy as np
```

### Intro to pyplot

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('Some Number')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

### Formatting the style of your plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
