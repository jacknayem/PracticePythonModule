```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([25, 30, 35, 40, 45, 50, 60, 75, 80, 100])
y = np.array([112, 120, 125, 130, 140, 145, 150, 155, 160, 170])
plt.plot(x, y)
plt.title('Sports Data')
plt.xlabel('Average Pulse')
plt.ylabel('Sports')
plt.show()
```

```python
x = np.array([25, 30, 35, 40, 45, 50, 60, 75, 80, 100])
y = np.array([112, 120, 125, 130, 140, 145, 150, 155, 160, 170])
labelsfont = {'family':'serif', 'color':'green', 'size':15}

plt.plot(x, y)
plt.title('Sports Data', fontdict ={'family':'serif','color':'blue', 'size':20})
plt.xlabel('Average Pulse', fontdict = labelsfont)
plt.ylabel('Sports', fontdict=labelsfont)
plt.show()
```

```python
x = np.array([25, 30, 35, 40, 45, 50, 60, 75, 80, 100])
y = np.array([112, 120, 125, 130, 140, 145, 150, 155, 160, 170])
labelsfont = {'family':'serif', 'color':'green', 'size':15}

plt.plot(x, y)
plt.title('Sports Data', fontdict ={'family':'serif','color':'blue', 'size':20}, loc='left')
plt.xlabel('Average Pulse', fontdict = labelsfont)
plt.ylabel('Sports', fontdict=labelsfont)
plt.show()
```

### Grid

```python
x = np.array([25, 30, 35, 40, 45, 50, 60, 75, 80, 100])
y = np.array([112, 120, 125, 130, 140, 145, 150, 155, 160, 170])
labelsfont = {'family':'serif', 'color':'green', 'size':15}

plt.plot(x, y, linestyle ='dashed')
plt.title('Sports Data', fontdict ={'family':'serif','color':'blue', 'size':20}, loc='left')
plt.xlabel('Average Pulse', fontdict = labelsfont)
plt.ylabel('Sports', fontdict=labelsfont)
plt.show()
```
