```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
x = np.array([0,6])
y = np.array([3,30])
plt.plot(x, y)
plt.show()
```

```python
x = np.array([0,6])
y = np.array([3,30])
plt.plot(x, y,'o')
plt.show()
```

```python
y = np.array([2,5,6,8])
plt.plot(y)
plt.show()
```

| Marker | Description

|'*'	| Star	
|'.'	| Point	
|','	| Pixel	
|'x'	| X	
|'X'	| X (filled)	
|'+'	| Plus	
|'P'	| Plus (filled)	
|'s'	| Square	
|'D'	| Diamond	
|'d'	| Diamond (thin)	
|'p'	| Pentagon	
|'H'	| Hexagon	
|'h'	|Hexagon

|'^'	| Triangle Up	
|'<'	| Triangle Left	
|'>'	| Triangle Right	
|'1'	| Tri Down	
|'2'	| Tri Up

|'4'	| Tri Right	
|'|'	| Vline	
| '_'	| Hline

```python
y = np.array([2,5,6,8])
plt.plot(y, marker='o')
plt.show()
```

```python
y = np.array([2,5,6,8])
plt.plot(y, 'o:r')
plt.show()
```

### Line Reference
| Line Syntax|Description|
|---|---|
|'-'|Solid line|
|':'|Dotted line|
|'--'|Dashed line|
|'-.'|Dashed/dotted line|

```python
y = np.array([2,5,6,8])
plt.plot(y, 'o-.')
plt.show()
```

### Color Reference
|Color Syntax | Description|
|---|---|
'r'	|Red	
'g'	|Green	
'b'	|Blue	
'c'	|Cyan	
'm'	|Magenta	
'y'	|Yellow	
'k'	|Black	
'w'	|White

```python
y = np.array([2,5,6,8])
plt.plot(y, 'oy')
plt.show()
```

```python
y = np.array([2,5,6,8])
plt.plot(y, 'or', ms=10) # ms = marker size
plt.show()
```

```python
y = np.array([2,5,6,8])
plt.plot(y, 'or', ms=10, mec='b') # mec = marker edge color
plt.show()
```

```python
y = np.array([2,5,6,8])
plt.plot(y, 'o', ms=10, mec='r', mfc='c') # mec = marker face color
plt.show()
```
