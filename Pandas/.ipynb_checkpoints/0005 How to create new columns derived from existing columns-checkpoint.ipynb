```python
import pandas as pd
```

```python
air_quality = pd.read_csv("data/air_quality_no2.csv")
air_quality.head()
```

```python
air_quality = air_quality[['CO', 'SO2','O3']]
air_quality.head()
```

```python
air_quality["new_column"] =  air_quality['CO'] * 2.345
air_quality.head()
```

```python
air_quality['new_column2'] = air_quality['CO']/ air_quality['SO2']
air_quality.head()
```

### Rename

```python
air_quality_rename = air_quality.rename(
    columns={
        'CO' : 'Carbon-Di-Oxide',
        'SO2': 'Salpuric-Acid',
        'O3': 'Tri-Oxide'
    }
)
air_quality_rename.head()
```

```python
air_quality_rename = air_quality_rename.rename(columns= str.lower)
air_quality_rename.head()
```
