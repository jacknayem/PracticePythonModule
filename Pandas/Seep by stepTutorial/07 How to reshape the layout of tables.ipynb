### Long to wide table format

```python
air_quality = pd.read_csv('data/air_quality_no2.csv')
```

```python
air_quality.head()
```

```python
no2 = air_quality[air_quality['City'] == 'Ahmedabad']
```

```python
no2_subset = no2.sort_index().groupby(['Date']).head(2)
no2_subset
```

```python
air_quality.pivot(columns='City', values='CO')
```

```python
air_quality.pivot_table(values='CO', index='City', aggfunc='mean')
```

```python
air_quality.pivot_table(values='NOx', index='City', aggfunc='mean', margins=True)
```

### Wide to long format

```python
no2_pivoted = no2.pivot(columns='City', values='NO2').reset_index()
no2_pivoted.head()
```

```python
no_2 = air_quality.melt(id_vars='City')
no_2.head()
```

```python
air_quality.melt(id_vars='City',
                value_vars=['NO', 'NO2', 'NOx'],
                )
```

### REMEMBER

* Sorting by one or more columns is supported by sort_values
* The pivot function is purely restructuring of the data, pivot_table supports aggregations
* The reverse of pivot (long to wide format) is melt (wide to long format)
